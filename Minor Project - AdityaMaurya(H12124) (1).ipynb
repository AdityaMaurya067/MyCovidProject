```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
df=pd.read_csv("Covid_19_india_june.csv")
df.info()
```

```python
df
```

```python
plt.figure(figsize=(10,5))
plt.plot(df['Date'], df['Confirmed'], label = "Confirmed")
plt.plot(df['Date'], df['Deaths'], label = "Deaths")
plt.plot(df['Date'], df['Cured'], label = "Cured")
plt.ylabel('Cases per day')
plt.title('COVID 19 CASES PER DAY IN DELHI DURING JUNE 2020')
plt.xticks(df['Date'],rotation=90)
plt.xticks(df.Date[0:30].tolist())
plt.legend(loc='best')
plt.show()
```

```python
from matplotlib import style
style.use('ggplot')
plt.figure(figsize=(10,5))
plt.grid()
plt.plot(df.Date,df.Confirmed,'y.-',label="Confirmed")
plt.plot(df.Date,df.Deaths,'r^-',label="Deaths")
plt.plot(df.Date,df.Cured,'g*-',label="Cured")
plt.title("Covid19 Pandemic Analysis of Delhi")
plt.xlabel("Date")
plt.ylabel("Confirmed cases")
plt.legend()
plt.xticks(df.Date[0:30:4].tolist())
plt.show()
```

```python
from matplotlib import pyplot as plt
ax = plt.axes()
ax.set_facecolor("dimgrey")
plt.figure(figsize=(20,10))
plt.plot(df.Date,df.Confirmed,color="red",label='Confirmed Cases')
plt.plot(df.Date,df.Cured,color="navy",label='Cured Cases')
plt.plot(df.Date,df.Deaths,color="orange",label='Death Cases')
plt.legend()
plt.bar(df.Date,df.Confirmed,label='Confirmed Cases',color='lightskyblue')
plt.bar(df.Date,df.Cured,color="palegreen",label='Cured Cases')
plt.bar(df.Date,df.Deaths,color="orange",label='Death Cases')
plt.title("Covid19 Pandemic Analysis of Delhi")
plt.xticks(df.Date,rotation=60)
plt.xlabel("Date",labelpad=10)
plt.ylabel("Number of Cases",labelpad=10)
plt.show()
```
